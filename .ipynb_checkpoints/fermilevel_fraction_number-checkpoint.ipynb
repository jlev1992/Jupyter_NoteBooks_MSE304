<p>
The density of the electron states in a metal can be written as $g(E) = AE^{1/2}$, where $A$ is a constant and $E$ is measured from the bottom of the conduction band. Derive the total number of states in this material and calculate the fraction of conduction electrons $f$ within $kT$ of the Fermi energy for {{params.element}} at 300 K. 
</p>
<p>
Hint: Use the approximation $\left(1-\frac{kT}{E_F}\right)^{3/2} \approx 1 - (3/2) (kT/E_F)$. Calculate the Fermi energy for this metal at 0 K using the expression for the density of electron states from lecture.
</p>

Solution: The expression for the density for a 3D electron gas is:

\begin{equation}
\frac{dn}{dE} = \frac{1}{2\pi^2} \left ( \frac{2m}{\hbar^2} \right )^{3/2} E^{1/2}
\end{equation}

\begin{equation}
n = \frac{1}{2\pi^2} \left ( \frac{2m}{\hbar^2} \right )^{3/2} \int_{0}^{E_f}E^{1/2}dE
\end{equation}


\begin{equation}
n = \frac{1}{2\pi^2} \left ( \frac{2m}{\hbar^2} \right )^{3/2} \frac{2}{3} E_f^{3/2}
\end{equation}


\begin{equation}
E_f = \left ( 3\pi^2 \left ( \frac{\hbar^2}{2m} \right )^{3/2} n \right )^{2/3}
\end{equation}

\begin{equation}
E_f =  \frac{\hbar^2}{2m} (3\pi^2)^{3/2} n^{2/3} = Cn^{2/3}
\end{equation}

Once you obtain $E_f$, you must calculate the fraction of electorns above $E_F$ at room temperature. You can make the assumption that the Fermi distribution remains 0K-like, so to find the total number of states you can do:

\begin{equation}
n_{total} = C\int_0^{E_F}E^{1/2}dE = C\frac{2}{3}E_{F}^{3/2}
\end{equation}

At finiate, small temperature $T$, electrons are excited within $kT$ of the Fermi surface. Thus, we can calculate the total number of electrons within kT of the fermi-surface at small T as the number between $E_{F}-KT$ and $E_F$:

\begin{equation}
n_{excited} = C\int_{E_F-kT}^{E_F}E^{1/2}dE = C\frac{3}{2}\left ( E_F^{3/2}- \left (E_F-kT \right )^{3/2} \right )
\end{equation}

Factoring out $E_F$ from the expression:

\begin{equation}
n_{excited} =  C\frac{3}{2} \left ( E_F^{3/2} -E_F^{3/2}  \left ( 1-\frac{kT}{E_F}\right )^{3/2} \right )
\end{equation}

Use the approximation:

\begin{equation}
\left ( 1-\frac{kT}{E_F}\right )^{3/2} \approx 1-\frac{3}{2}kTE^{1/2}
\end{equation}

Plugging this back in, we find:

\begin{equation}
n_{excited} = CkTE^{1/2}
\end{equation}

Taking the ratio to get the fraction of electrons in the valence states:

\begin{equation}
f_{excited} = \frac{n_{excited}}{n_{total}} = \frac{3}{2}\frac{kT}{E_F}
\end{equation}
