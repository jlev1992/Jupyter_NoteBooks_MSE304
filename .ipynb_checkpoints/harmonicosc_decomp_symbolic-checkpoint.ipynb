Problem:

<p>
If you find the eigenvalues of the matrix from the previous question,
you can find new coordinates $x_s$ and $x_c$ that describe the problem
in terms of the center of mass of both oscillators and the separation
of both oscillators. Both look like springs and are independent:
</p>
<p>
$\hat{H} = \hat{H}_c + \hat{H}_s$,
</p>
<p>
where
</p>
<p>
$\hat{H}_c = \frac{\hat{p}_c^2}{2m} + \left(C-\frac{2K}{R^3}\right)x_c^2$,
</p>
<p>
$\hat{H}_s = \frac{\hat{p}_s^2}{2m} + \left(C+\frac{2K}{R^3}\right)x_s^2$.
</p>
<p>
What are the frequencies of each mode?
</p>

</p>
<p>

Solution:

We know that the usual expression for the Hamiltonian for a single particle in a harmonic potential is:

\begin{equation}
H = \frac{p^2}{2m}+\frac{1}{2}m\omega^2x^2
\end{equation}

We can rewrite the the potential terms:

\begin{equation}
\left(C-\frac{2K}{R^3}\right)x_c^2 = \frac{1}{2}m\left(\frac{2C}{m}-\frac{4K}{mR^3}\right)x_c^2
\end{equation}

\begin{equation}
\left(C+\frac{2K}{R^3}\right)x_2^2 = \frac{1}{2}m\left(\frac{2C}{m}+\frac{4K}{mR^3}\right)x_s^2
\end{equation}

And we see now that the two frequencies are:

\begin{equation}
\omega_c = \sqrt{\left(\frac{2C}{m}-\frac{4K}{mR^3}\right)}
\end{equation}

\begin{equation}
\omega_s = \sqrt{\left(\frac{2C}{m}+\frac{4K}{mR^3}\right)}
\end{equation}
