Solution:

All terms that rely on the interaction between the systems (seperation $R$ and change in frequency $\Delta \omega$) go to 0. So, the expression for the non-interacting ground state energy is now:

\begin{equation}
E_g = \frac{\hbar}{2}\left(\sqrt{\frac{C}{m}}+\sqrt{\frac{C}{m}}\right) = \hbar \sqrt{\frac{C}{m}}
\end{equation}
