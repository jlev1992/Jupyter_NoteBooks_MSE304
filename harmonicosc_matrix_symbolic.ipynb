Given out Hamilotnian now is:

\begin{equation}
H = H_1+H_2+H_C = \frac{p_1^2}{2m}+\frac{1}{2}Cx_1^2 + \frac{p_2^2}{2m}+\frac{1}{2}Cx_2^2 -\frac{2x_1x_2K}{R^3}
\end{equation}

We need to seperate this into the momentum matrix product and the harmonic optantial product:

\begin{equation}
H = \left ( \frac{p_1^2}{2m} + \frac{p_2^2}{2m} \right ) + \left ( \frac{1}{2}Cx_1^2 + \frac{1}{2}Cx_2^2  -\frac{2x_1x_2K}{R^3} \right )
\end{equation}

We can tell right away that the terms with $C$ will be on the diagonal, since they have $x_n^2$ terms attached. The $K/R^3$ terms will be on the off diagonal, since they have products of $x_1$ and $x_2$. After playing with the nmumbers, it is straight forward to determine:


\begin{equation}
a = \frac{C}{2}
\end{equation}

\begin{equation}
b = -\frac{K}{R^3}
\end{equation}
